### Gradient Descent Algorithm

    It's an optimization technique to calculate parameters by reducing a given cost function

an iterative algorithm to estimate best parameters using training data and cost function

1. Randomly initialize weights/parameters
2. take gradient of cost function with respect to weights / parameters
3. update the value of parameters by subtracting eta*gradients(Descent) from parameters
4. Repeat step 2 and 3 until algorithm converge

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(121)
```

```python
m = 100
X1 = np.random.random((m,1))
y = 10 + 5*X1 + np.random.random((m,1))*2
# y = b + w1 * X1 + noise
plt.scatter(X1, y)
plt.show()
```

```python
# step1 -> select random weights / parameters
w1 = np.random.random()
b = np.random.random()
print("w1 = ",w1," \n","b = ",b)

y_hat = b + w1*X1

plt.scatter(X1, y, label="original")
plt.plot(X1, y_hat, label="prediction", color="red")
plt.xlabel("X1")
plt.ylabel("y")
plt.show()
```

```python
# step take gradient of w and b
eta = 0.1 # learning rate
gw1 = (2/m)*((b+w1*X1 - y)*X1).sum()
gb = (2/m)*((b+w1*X1 - y)).sum()
print(f"gw: {gw1} \ngb: {gb}")

# update the value of w , and b by subtracting gradient
w1 = w1 - (eta * gw1)
b = b - (eta*gb)
print("Gradient Descent Step")
print(f"w1 = {w1} \n b = {b}")

y_hat = b + w1*X1
plt.scatter(X1, y, label="original")
plt.plot(X1, y_hat, label="prediction", color = 'r')
plt.xlabel("X1")
plt.ylabel("y")
plt.show()
```

```python
# Random initialization

w1 = np.random.random()
b = np.random.random()

eta = 0.1 # learning rate
epoch = 100 # number of times we repeat it

y_hat = b + w1*X1 # initial prediction
rmse = np.sqrt(( (y_hat - y)**2 ).sum()/m)
print(f"w = {w1:<5.2f}  b = {b:<5.2f}  rmse = {rmse:<5.2f}")
plt.scatter(X1, y, label="original")
plt.plot(X1, y_hat, label="prediction", color = 'r')
plt.xlabel("X1")
plt.ylabel("y")

for _ in range(epoch):
# gradient calculation
    gw1 = (2/m)*((b+w1*X1 - y)*X1).sum()
    gb = (2/m)*((b+w1*X1 - y)).sum()
# Descent to minimize error
    w1 = w1 - (eta*gw1)
    b = b - (eta*gb)
    y_hat = b + w1*X1
    rmse = np.sqrt(((y_hat - y)**2).sum()/m)
    plt.plot(X1, y_hat, color="red")
    print(f"w = {w1.round(2):<5}  b = {b.round(2):<5}  rmse = {rmse.round(2):<5}")

plt.show()
```

### Gradient Descent

$\hat y = X.\theta$

$MSE(\theta) = \frac {1} {m} \sum_{i=0}^m {(X^{(i)}\theta - y^{(i)})}^2$

$\nabla_{\theta}MSE{\theta} = \frac {2} {m} X^T(X\theta - y)$

#### Gradient Descent Step 

$\theta^{(next)} = \theta - \eta \nabla_{\theta}MSE{\theta} $

```python
np.random.seed(123)
m = 100 # number of instances

X = np.random.random((m,2))

y = (20 + 2*X[:,0] + 8*X[:,1] + np.random.random(m)).reshape(-1,1)
```

### Batch Gradient Descent

$\nabla_{\theta}MSE{\theta} = \frac {2} {m} X^T(X\theta - y)$

```python
import time
```

```python
eta = 0.1 # learning rate
n_epochs = 500 # no. of iterations
tol = 0.001 # early stopping

X_b = np.c_[np.ones((m,1)), X]

# random initialization of weights / thetas
thetas = np.random.random((3,1)) # randomly initialize model parameters
errors = []

for epoch in range(n_epochs):
    gradients = 2/m*(X_b.T)@(X_b @ thetas - y)
    # (3, m) @ ((m, 3)*(3, 1) - (m, 1)
    # gradient descent
    y_hat_prev = X_b @ thetas
    prev_error = ((y_hat_prev - y)**2).sum()/m
    thetas = thetas - eta * gradients
    y_hat_current = X_b @ thetas
    current_error = ((y_hat_current - y)**2).sum()/m
    print(current_error.round(2), end="\t")
    errors.append(current_error)
    if abs(current_error - prev_error)<tol:
        # early stopping
        print(f"\nStopping early at epoch {epoch}")
        break

    #time.sleep(1)
```

```python
plt.plot(errors)
plt.yscale("log")
plt.show()
```
