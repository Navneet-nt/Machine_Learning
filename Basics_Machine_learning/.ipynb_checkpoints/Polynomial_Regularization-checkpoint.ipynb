```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(123)
m = 100
X = 9*np.random.random((m,1))-6
y = 0.4 * X**2 + 2.5 * X + 20 +np.random.random((m,1))*5
y = y.ravel()
# y = w2*x^2 + w1*x + b
```

```python
plt.scatter(X,y)
plt.xlabel("X")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
poly = PolynomialFeatures(degree=2, include_bias = False)
```

```python
poly.fit(X)
```

```python
X_tr = poly.transform(X)
```

```python
X_tr[:5]
```

```python
poly_reg  = make_pipeline(
    PolynomialFeatures(degree=2, include_bias=False),
    StandardScaler(),
    SGDRegressor()
)
```

```python
poly_reg.fit(X, y)
```

```python
poly_reg.steps[-1][1].intercept_
```

```python
poly_reg.steps[-1][1].coef_
```

```python
# y = 0.4 * X**2 + 2.5 * X + 20 +np.random.random((m,1))*5
```

```python
y_hat = poly_reg.predict(X)
```

```python
X.min() , X.max()
```

```python
plt.scatter(X, y)
plt.scatter(X, y_hat)
```

```python
X_new = np.linspace(-6, 3, 100).reshape(-1,1)
y_hat = poly_reg.predict(X_new)
```

```python
plt.scatter(X,y)
plt.plot(X_new, y_hat, color='red', lw=2)
plt.show()
```

```python
poly_reg  = make_pipeline(
    PolynomialFeatures(degree=2, include_bias=False),
    StandardScaler(),
    SGDRegressor()
)

poly_reg.fit(X, y)

X_new = np.linspace(-6, 3, 100).reshape(-1,1)
y_hat = poly_reg.predict(X_new)

plt.scatter(X,y)
plt.plot(X_new, y_hat, color='red', lw=2)
plt.show()
```

```python
from warnings import filterwarnings
filterwarnings(action='ignore')
```

```python
from sklearn.model_selection import learning_curve
```

```python
train_sizes, train_scores, valid_scores = learning_curve(poly_reg, X, y,
                                                     train_sizes=np.linspace(0.01, 1.0, 40),
                                                     cv = 5, scoring='neg_mean_squared_error')
```

```python
train_scores = np.sqrt(-train_scores)
valid_scores = np.sqrt(-valid_scores)
```

```python
train_scores = train_scores.mean(axis=1) # because of cross validation
# we were getting 5 error for each step
valid_scores = valid_scores.mean(axis=1)
```

```python
print(train_sizes[:5])
print()
print(train_scores[:5])
print()
print(valid_scores[:5])
```

```python
plt.plot(train_sizes, train_scores, "r--o",lw=2,label="Training Error")
plt.plot(train_sizes, valid_scores, "b:+",lw=2,label="Validation Error")

plt.grid()
plt.legend()
plt.show()
```

```python
poly_reg  = make_pipeline(
    PolynomialFeatures(degree=2, include_bias=False),
    StandardScaler(),
    SGDRegressor()
)

poly_reg.fit(X, y)

train_sizes, train_scores, valid_scores = learning_curve(poly_reg, X, y,
                                                     train_sizes=np.linspace(0.01, 1.0, 40),
                                                cv = 5, scoring='neg_mean_squared_error')

train_scores = np.sqrt(-train_scores)
valid_scores = np.sqrt(-valid_scores)

train_scores = train_scores.mean(axis=1) # because of cross validation
# we were getting 5 error for each step
valid_scores = valid_scores.mean(axis=1)

plt.plot(train_sizes, train_scores, "r--o",lw=2,label="Training Error")
plt.plot(train_sizes, valid_scores, "b:+",lw=2,label="Validation Error")
plt.grid()
plt.title("Learning Curve")
plt.legend()
plt.show()
```
