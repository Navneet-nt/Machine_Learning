```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### scatter plot

-> to find out relationship between two variables

```python
tips = sns.load_dataset("tips")
tips
```

```python
tips.shape
```

```python
tips.head()
```

```python
#line plot
plt.plot("total_bill","tip",data=tips,alpha=0.8)
plt.show()
```

```python
# scatter plot
plt.plot("total_bill","tip",'o',data=tips,alpha=0.5)
plt.show()
```

```python
plt.plot(tips.total_bill,tips.tip,"H",alpha=0.5,ms=20)
plt.show()
```

```python
plt.scatter("total_bill","tip",data=tips,marker="h",alpha=0.6)
plt.show()
```

```python
tips["sex"].unique()
```

```python
c = tips["sex"].map({"Female":'pink','Male':'yellow'}).to_numpy()
c
```

```python
tips["smoker"].unique()
```

```python
w = tips["smoker"].map({"Yes":"red","No":"green"}).to_numpy()
w
```

```python
f = np.append(c,w)
g = np.random.choice(f,size=244,replace=False)
g
```

```python
plt.scatter("total_bill","tip",data=tips,marker="h",c=g,s=100,alpha=0.5)
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title("Relation between tips paid by gender")
plt.grid()
#to show labels using legend()
plt.plot([],[],'h',color='pink',label='Female')
plt.plot([],[],'h',color='yellow',label='Male')
plt.plot([],[],'h',color='red',label='smokers')
plt.plot([],[],'h',color='green',label='non-smokers')
plt.legend()
plt.show()
```

```python
sns.set_style("whitegrid")
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex",size="smoker",style="size",palette="coolwarm")
plt.show()
```

```python
colors = ['red','green','blue','cyan','magenta','pink','purple','gold','silver']
```

```python
freq,bins,objs = plt.hist("total_bill",data=tips,bins=20) #static plot
for obj in objs:
    obj.set_facecolor(np.random.choice(colors))
```

```python
#fig,(ax1,ax2) = plt.subplots()
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(8,4),sharey=True)

ax1.hist(tips[tips['sex']=='Male']['total_bill'],bins=20,color='red',ec='k')
ax2.hist(tips[tips['sex']=='Female']['total_bill'],bins=20,color='yellow',ec='k')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',hue='sex',rug=True,kind='kde')
plt.show()
```

#### Box Plot

```python
objs = plt.boxplot("total_bill",data=tips,showmeans=True,patch_artist=True)

for obj in objs["caps"]:
    obj.set_linewidth(5)
    obj.set_color("green")
    
for obj in objs['whiskers']:
    obj.set_color("red")
    obj.set_linestyle('-')
    obj.set_linewidth(2)

for obj in objs["fliers"]:
    obj.set_marker("D")
    obj.set_markersize(8)
    obj.set_markerfacecolor("pink")
```

```python
sns.boxplot(data=tips,x='sex',y='total_bill',hue='smoker')
plt.show()
```

```python
sns.boxenplot(data=tips,x='sex',y='total_bill',hue='smoker')
plt.show()
```

```python
#category plot -> we can create any kind of plot
# by default it creates 'strip' plot
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip')
#scatter with histogram
plt.show()
```

```python
sns.pairplot(tips)
#tips -> size,total_bill,tip
#tip tip -> histogram,
plt.show()
```

```python
d = tips.groupby("day")["total_bill"].mean().reset_index()
d
```

```python
plt.bar('day','total_bill',data=d,color=['r','g','b','c'])
plt.show()
```

```python
plt.pie(d['total_bill'],labels=d['day'],labeldistance=0.5,autopct='%.2f%%',pctdistance=1.2,
       explode=[0,0,0,0.1],colors=['grey','grey','grey','red'],shadow=True,radius=1)
plt.show()
```

```python
from matplotlib.patches import Circle
```

```python
a = Circle((0,0),0.2,facecolor='white')
```

```python
# donut plot
plt.pie(d['total_bill'],labels=d['day'],labeldistance=0.5,autopct='%.2f%%',pctdistance=1.2,
       explode=[0,0,0,0.1],colors=['grey','grey','grey','red'],shadow=True,radius=1)
plt.gca().add_patch(a)
plt.show()
```
