```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
df = sns.load_dataset("flights")
df
```

```python
ans = df.groupby("year")["passengers"].mean().reset_index()
ans
```

```python
fig = plt.figure(figsize=(6,4),facecolor="gold",dpi=100)
axes = fig.add_axes([0,0,1,1])
axes.plot("year","passengers",ls='--',marker='o',mfc='red',ms=10,mec='black',data=ans,label="data")
axes.set_title("passengers over the years")
axes.set_xlabel("Year")
axes.set_ylabel("no of passengers")
axes.grid()
axes.legend()
plt.show()
```

```python
new_df = df.pivot_table(columns='year',index='month',values='passengers')
```

```python
new_df
```

```python
plt.plot(new_df[1959],ls='--',marker='o',mfc='r',ms=10,color='k',mec='y',label='1959')
plt.plot(new_df[1960],ls=':',marker='s',mfc='b',ms=10,color='g',mec='k',label='1960')
plt.legend(loc='upper left',edgecolor='k')
plt.title("Trend of Passengers over the year",color='red',
          fontdict={"fontsize":10,"family":"cursive","weight":0})
axes = plt.gca() # get current axes
axes.spines["right"].set_visible(False)
axes.spines["top"].set_visible(False)
axes.spines["left"].set_linestyle(":")
# axes.set_xticklabels(xticks,labels)
#first set ticks then go for labels
axes.set_xticks(new_df.index)
axes.set_xticklabels(new_df.index,color='r')

yticks = np.arange(300,700,25)
axes.set_yticks(yticks)
axes.set_yticklabels([f"{chr(65+i)}-{yticks[i]}" for i in range(16)],color='b')
#set_yticks(ticks,tick_labels)
#set_yticklabels(tick_labels) #make sure to set the ticks

axes.grid(ls=':',color='m',alpha=0.8)
axes.text(1,575,"Navneet Tiwari",color='r',fontsize=15,alpha=0.7)

from matplotlib.patches import Rectangle
rct = Rectangle((0.5,575),5,30,edgecolor='k',alpha=0.2,facecolor='w')
axes.add_patch(rct)
plt.show()
```
